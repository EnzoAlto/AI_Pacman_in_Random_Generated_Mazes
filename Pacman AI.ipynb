```python
import pygame, sys, os

# ---------- MAP Editor -----------------------------------
MAP_TEXT = """
####################
#........P.........#
#.#######..#######.#
#..o..........o....#
#.#######..#######.#
#..................#
#.#######..#######.#
#.#..G.......G...#.#
#.#.#####..#####.#.#
#....o.......o.....#
####################
""".strip("\n")

# --- Config ----------------------------------------------------------------------------------
TILE_SIZE = 30
FPS = 60
GHOST_SPRITE = 'ghost.png'

# Tile codes
EMPTY, WALL, PELLET, POWER = 0, 1, 2, 3

# Setting Colors for each element of the game
BLACK   = (0, 0, 0)
WALL_C  = (0, 0, 255)
PELLET_C= (255, 255, 0)
POWER_C = (255,184,151)
PACMAN_C= (255, 255, 0)

# Defining How the map generator recognizes the element
CHAR2TILE = {'#': WALL, '.': PELLET, ' ': EMPTY, 'o':POWER}

# --- Map parsing -----------------------------------------------------------------


def load_map_from_text(txt):
    lines = [line.rstrip('\n') for line in txt.splitlines()]
    rows, cols = len(lines), max(len(r) for r in lines)
    lines = [r.ljust(cols, ' ') for r in lines]

    grid = [[EMPTY]*cols for _ in range(rows)]
    pac_spawn = None
    ghost_spawns = []

    for r, row in enumerate(lines):
        for c, ch in enumerate(row):
            if ch in CHAR2TILE:
                grid[r][c] = CHAR2TILE[ch]
            elif ch == 'P':
                pac_spawn = (r, c)
            elif ch == 'G':
                ghost_spawns.append((r, c))

    # If no pellets, auto-fill empties
    if not any(PELLET in row for row in grid):
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == EMPTY:
                    grid[r][c] = PELLET

    if pac_spawn is None:
        pac_spawn = (1, 1)
    if not ghost_spawns:
        ghost_spawns = [(rows-2, cols-2)]

    return grid, pac_spawn, ghost_spawns
# ---- Utilities ----------------------------------------
def manhattan(a_r, a_c, b_r, b_c):
    return abs(a_r - b_r) + abs(a_c - b_c)

def clamp(v, lo, hi):
    return max(lo, min(hi, v))
# --- Entities ------------------------------------------------
class Pacman:
    def __init__(self, r, c):
        self.x = (c + 0.5) * TILE_SIZE
        self.y = (r + 0.5) * TILE_SIZE
        self.dir_x = 0
        self.dir_y = 0
        self.next_dx = 0
        self.next_dy = 0
        self.speed = 2.0  # divides TILE_SIZE (20 % 2 == 0)
        self.radius = TILE_SIZE // 2 - 2
        self._EPS_CENTER = 0.5

    def tile_pos(self):
        return int(self.y // TILE_SIZE), int(self.x // TILE_SIZE)

    def _center_of(self, r, c):
        return (c + 0.5) * TILE_SIZE, (r + 0.5) * TILE_SIZE

    def _is_centered(self):
        cx = (self.x % TILE_SIZE) - TILE_SIZE/2
        cy = (self.y % TILE_SIZE) - TILE_SIZE/2
        return abs(cx) <= self._EPS_CENTER and abs(cy) <= self._EPS_CENTER

    def _snap_to_center(self):
        r, c = self.tile_pos()
        self.x, self.y = self._center_of(r, c)

    def _can_move(self, grid, dx, dy):
        if dx == 0 and dy == 0:
            return True
        rows, cols = len(grid), len(grid[0])
        r, c = self.tile_pos()
        nr, nc = r + dy, c + dx
        return 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] != WALL

    def set_desired_dir(self, dx, dy, grid):
        self.next_dx, self.next_dy = dx, dy
        if self._is_centered() and self._can_move(grid, dx, dy):
            self.dir_x, self.dir_y = dx, dy

    def _advance_towards_next_center(self):
        if self.dir_x == 0 and self.dir_y == 0:
            return
        r, c = self.tile_pos()
        target_r = r + self.dir_y
        target_c = c + self.dir_x
        tx, ty = self._center_of(target_r, target_c)
        step = self.speed
        # snap if reaching center this frame
        if (tx - self.x)**2 + (ty - self.y)**2 <= step**2:
            self.x, self.y = tx, ty
            return
        if self.dir_x != 0:
            self.x = min(self.x + step, tx) if self.dir_x > 0 else max(self.x - step, tx)
            self.y = (r + 0.5) * TILE_SIZE
        else:
            self.y = min(self.y + step, ty) if self.dir_y > 0 else max(self.y - step, ty)
            self.x = (c + 0.5) * TILE_SIZE

    def update(self, grid):
        # if self._is_centered():
        #     self._snap_to_center()
        #     if (self.next_dx, self.next_dy) != (self.dir_x, self.dir_y):
        #         if self._can_move(grid, self.next_dx, self.next_dy):
        #             self.dir_x, self.dir_y = self.next_dx, self.next_dy
        #     if not self._can_move(grid, self.dir_x, self.dir_y):
        #         self.dir_x = self.dir_y = 0
        # self._advance_towards_next_center()
        # if self._is_centered():
        #     self._snap_to_center()
        #     r, c = self.tile_pos()
        #     if grid[r][c] == PELLET:
        #         grid[r][c] = EMPTY
        eaten = None
        if self._is_centered():
            self._snap_to_center()
            if (self.next_dx, self.next_dy) != (self.dir_x, self.dir_y):
                if self._can_move(grid, self.next_dx, self.next_dy):
                    self.dir_x, self.dir_y = self.next_dx, self.next_dy
            if not self._can_move(grid, self.dir_x, self.dir_y):
                self.dir_x = self.dir_y = 0
        self._advance_towards_next_center()

        # Checks if Pacman has eaten a regular pellet or a powerup
        if self._is_centered():
            self._snap_to_center()
            r, c = self.tile_pos()
            if grid[r][c] == PELLET:
                grid[r][c] = EMPTY
                eaten = 'pellet'
            elif grid[r][c] == POWER:
                grid[r][c] = EMPTY
                eaten = 'power'
        return eaten
        

    def draw(self, screen):
        pygame.draw.circle(screen, PACMAN_C, (int(self.x), int(self.y)), self.radius)

class Ghost:
    def __init__(self, r, c, sprite):
        self.x = (c + 0.5) * TILE_SIZE
        self.y = (r + 0.5) * TILE_SIZE
        self.dir_x = 0
        self.dir_y = 0
        self.speed = 2.0
        self.sprite = sprite
        self._EPS_CENTER = 0.5
        self.spawn_r, self.spawn_c = r, c # Remember home when spawned
        self.frightened = False
    
    def _is_centered(self):
        cx = (self.x % TILE_SIZE) - TILE_SIZE/2
        cy = (self.y % TILE_SIZE) - TILE_SIZE/2
        return abs(cx) <= self._EPS_CENTER and abs(cy) <= self._EPS_CENTER

    def _snap_to_center(self):
        r, c = int(self.y // TILE_SIZE), int(self.x // TILE_SIZE)
        self.x = (c + 0.5) * TILE_SIZE
        self.y = (r + 0.5) * TILE_SIZE

    def _can_move(self, grid, dx, dy):
        rows, cols = len(grid), len(grid[0])
        r, c = int(self.y // TILE_SIZE), int(self.x // TILE_SIZE)
        nr, nc = r + dy, c + dx
        return 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] != WALL

    def _choose_dir_towards(self, grid, pac_r, pac_c): # running towards pacman
        col = int(self.x // TILE_SIZE)
        row = int(self.y // TILE_SIZE)
        best_d, best = 1e9, (0,0)
        for dx, dy in [(0,-1),(1,0),(0,1),(-1,0)]:
            if self._can_move(grid, dx, dy):
                nr, nc = row + dy, col + dx
                d = abs(nc - pac_c) + abs(nr - pac_r)
                if d < best_d:
                    best_d, best = d, (dx, dy)
        self.dir_x, self.dir_y = best

    def _choose_dir_away(self, grid, pac_r, pac_c): # running away from pacman
        col = int(self.x // TILE_SIZE)
        row = int(self.y // TILE_SIZE)
        best_d, best = -1, (0,0)
        for dx, dy in [(0,-1),(1,0),(0,1),(-1,0)]:
            if self._can_move(grid, dx, dy):
                nr, nc = row + dy, col + dx
                d = abs(nc - pac_c) + abs(nr - pac_r)
                if d > best_d:
                    best_d, best = d, (dx, dy)
        self.dir_x, self.dir_y = best


    def _advance_towards_next_center(self):
        if self.dir_x == 0 and self.dir_y == 0:
            return
        r, c = int(self.y // TILE_SIZE), int(self.x // TILE_SIZE)
        tx = (c + 0.5 + self.dir_x) * TILE_SIZE
        ty = (r + 0.5 + self.dir_y) * TILE_SIZE
        step = self.speed
        if (tx - self.x)**2 + (ty - self.y)**2 <= step**2:
            self.x, self.y = tx, ty
            return
        if self.dir_x != 0:
            self.x = min(self.x + step, tx) if self.dir_x > 0 else max(self.x - step, tx)
            self.y = (r + 0.5) * TILE_SIZE
        else:
            self.y = min(self.y + step, ty) if self.dir_y > 0 else max(self.y - step, ty)
            self.x = (c + 0.5) * TILE_SIZE

    def update(self, grid, pac_r, pac_c):
        if self._is_centered():
            self._snap_to_center()
            if self.frightened: 
                self._choose_dir_away(grid, pac_r, pac_c)
                self.speed = 1.5 # Speed for Ghost running away from pacman
            else:
                self._choose_dir_towards(grid, pac_r, pac_c)
                self.speed = 2.0 # Speed of ghost chasing pacman
            if not self._can_move(grid, self.dir_x, self.dir_y):
                self.dir_x = self.dir_y = 0 # determines if ghosts bumps a wall or not
        self._advance_towards_next_center()

    def respawn_to_spawn(self): # Sets Ghost back to its spawn point
        self.x = (self.spawn_c + 0.5) * TILE_SIZE
        self.y = (self.spawn_r + 0.5) * TILE_SIZE
        self.dir_x = self.dir_y = 0

    def draw(self, screen):
        if self.frightened or not (self.sprite and hasattr(self.sprite, "get_rect")):
            color = (100,100,255) if self.frightened else (255,0,0) # Color Change when Scared or not
            pygame.draw.circle(screen, color, (int(self.x), int(self.y)), TILE_SIZE//2 - 2)
        else:
            rect = self.sprite.get_rect(center=(int(self.x), int(self.y)))
            screen.blit(self.sprite, rect) # not sure yet how this works

# --- Setup & main loop --------------------------------------
def make_game_from_text(txt):
    grid, pac_spawn, ghost_spawns = load_map_from_text(txt)
    rows, cols = len(grid), len(grid[0])
    width, height = cols*TILE_SIZE, rows*TILE_SIZE

    pac = Pacman(*pac_spawn)
    ghost_sprite = None
    if os.path.exists(GHOST_SPRITE):
        try:
            ghost_sprite = pygame.image.load(GHOST_SPRITE).convert_alpha()
            ghost_sprite = pygame.transform.scale(ghost_sprite, (TILE_SIZE, TILE_SIZE))
        except Exception:
            ghost_sprite = None
    ghosts = [Ghost(r, c, ghost_sprite) for (r, c) in ghost_spawns]
    return grid, pac, ghosts, rows, cols, width, height


def count_pellets(g):
    return sum(cell == PELLET for row in g for cell in row)



def main():
    pygame.init()
    grid, pac, ghosts, ROWS, COLS, WIDTH, HEIGHT = make_game_from_text(MAP_TEXT)
    screen = pygame.display.set_mode((WIDTH, HEIGHT))
    pygame.display.set_caption("Pac-Man (inline map, simple)")
    clock = pygame.time.Clock()
    font = pygame.font.SysFont(None, 18)

    pellets_left = count_pellets(grid)

    # --- NEW: frightened state ---
    FRIGHTENED_SECS = 6
    FRIGHTENED_FRAMES = int(FPS * FRIGHTENED_SECS)
    frightened_timer = 0

    def reset_game():
        nonlocal grid, pac, ghosts, pellets_left, ROWS, COLS, WIDTH, HEIGHT, frightened_timer
        grid, pac_spawn, ghost_spawns = load_map_from_text(MAP_TEXT)
        pac = Pacman(*pac_spawn)
        spr = ghosts[0].sprite if ghosts else None
        ghosts = [Ghost(r, c, spr) for (r, c) in ghost_spawns]
        ROWS, COLS = len(grid), len(grid[0])
        WIDTH, HEIGHT = COLS*TILE_SIZE, ROWS*TILE_SIZE
        pellets_left = count_pellets(grid)
        frightened_timer = 0  # clear state

    running = True
    while running:
        clock.tick(FPS)
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_UP:    pac.set_desired_dir(0, -1, grid)
                if event.key == pygame.K_DOWN:  pac.set_desired_dir(0,  1, grid)
                if event.key == pygame.K_LEFT:  pac.set_desired_dir(-1, 0, grid)
                if event.key == pygame.K_RIGHT: pac.set_desired_dir(1,  0, grid)
                if event.key == pygame.K_r:     reset_game()

        # Pac-Man update + consume
        eaten = pac.update(grid)
        if eaten == 'power':
            frightened_timer = FRIGHTENED_FRAMES
            for g in ghosts:
                g.frightened = True

        pr, pc = pac.tile_pos()

        # Decrement frightened timer
        if frightened_timer > 0:
            frightened_timer -= 1
            if frightened_timer == 0:
                for g in ghosts:
                    g.frightened = False

        # Ghosts update + collisions
        for g in ghosts:
            g.update(grid, pr, pc)
            gr, gc = int(g.y//TILE_SIZE), int(g.x//TILE_SIZE)
            if (gr, gc) == (pr, pc):
                if g.frightened:
                    g.respawn_to_spawn()
                    g.frightened = False   # <--- eaten ghost reverts to red
                    continue
                else:
                    reset_game()
                    break

        # Win check (unchanged)
        pellets_left = count_pellets(grid)
        if pellets_left == 0:
            for r in range(ROWS):
                for c in range(COLS):
                    if grid[r][c] == EMPTY:
                        grid[r][c] = PELLET
            pellets_left = count_pellets(grid)
            reset_game()

        # Draw
        screen.fill(BLACK)
        for r in range(ROWS):
            for c in range(COLS):
                cell = grid[r][c]
                if cell == WALL:
                    pygame.draw.rect(screen, WALL_C, (c*TILE_SIZE, r*TILE_SIZE, TILE_SIZE, TILE_SIZE))
                elif cell == PELLET:
                    pygame.draw.circle(screen, PELLET_C, (c*TILE_SIZE + TILE_SIZE//2, r*TILE_SIZE + TILE_SIZE//2), 3)
                elif cell == POWER:
                    pygame.draw.circle(screen, POWER_C, (c*TILE_SIZE + TILE_SIZE//2, r*TILE_SIZE + TILE_SIZE//2), 7)

        pac.draw(screen)
        for g in ghosts:
            g.draw(screen)

        frightened_any = any(g.frightened for g in ghosts)

        text = font.render(
            f"Pellets: {pellets_left}    Power: {'ON' if frightened_any else 'off'}    [Arrows] Move  [R] Reset",
            True, (200,200,200)
        )
        screen.blit(text, (6, 4))
        pygame.display.flip()

    pygame.quit(); sys.exit()


if __name__ == "__main__":
    main()
```

```python
# pacman_inline_map.py
# ------------------------------------------------------------
# Arrows: move (when AI off)     A: toggle AI on/off
# E: edit mode on/off            1/2/3: paint Wall/Pellet/Empty
# P: set Pac-Man spawn           G: add Ghost spawn
# R: reset (keeps map)           S: print current map to console
# ------------------------------------------------------------

import pygame, sys, os
from collections import deque
import random

# =========================
# Inline ASCII Map (edit me)
# =========================
# MAP_TEXT = """
# ####################
# #........P.........#
# #....###.....###...#
# #..................#
# #..................#
# #....###.....###...#
# #..................#
# #..................#
# #....###.....###...#
# #....G.........G...#
# #..................#
# #..................#
# ####################
# """.strip("\n")

MAP_TEXT = """
####################
#........P.........#
#....###.....###...#
#..................#
#..................#
#....###.....###...#
#..................#
#..................#
#....###.....###...#
#..................#
#..................#
#..................#
####################
""".strip("\n")

# ------------- Config -------------
TILE_SIZE = 20
FPS = 60
GHOST_SPRITE = 'ghost.png'

# Tile codes
EMPTY, WALL, PELLET = 0, 1, 2

# Colours
BLACK   = (0, 0, 0)
WALL_C  = (0, 0, 255)
PELLET_C= (255, 255, 0)
PACMAN_C= (255, 255, 0)
GRID_C  = (40, 40, 40)
EDIT_HI = (60, 60, 60)

CHAR2TILE = {'#': WALL, '.': PELLET, ' ': EMPTY}
TILE2CHAR = {v:k for k,v in CHAR2TILE.items()}

# ------------- Map Parse / Print -------------
def load_map_from_text(txt):
    lines = [line.rstrip('\n') for line in txt.splitlines()]
    rows = len(lines)
    cols = max(len(r) for r in lines)
    lines = [r.ljust(cols, ' ') for r in lines]

    grid = [[EMPTY]*cols for _ in range(rows)]
    pac_spawn = None
    ghost_spawns = []

    for r, row in enumerate(lines):
        for c, ch in enumerate(row):
            if ch in CHAR2TILE:
                grid[r][c] = CHAR2TILE[ch]
            elif ch == 'P':
                pac_spawn = (r, c)
            elif ch == 'G':
                ghost_spawns.append((r, c))
            else:
                grid[r][c] = EMPTY

    # If no pellets were in the text, auto-fill pellets on empties
    if not any(PELLET in row for row in grid):
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == EMPTY:
                    grid[r][c] = PELLET

    if pac_spawn is None:
        pac_spawn = (1, 1)
    if not ghost_spawns:
        ghost_spawns = [(rows-2, cols-2)]

    return grid, pac_spawn, ghost_spawns

def stringify_map(grid, pac_spawn, ghost_spawns):
    rows, cols = len(grid), len(grid[0])
    out = [[' ']*cols for _ in range(rows)]
    for r in range(rows):
        for c in range(cols):
            out[r][c] = TILE2CHAR.get(grid[r][c], ' ')
    pr, pc = pac_spawn
    out[pr][pc] = 'P'
    for gr, gc in ghost_spawns:
        out[gr][gc] = 'G'
    return "\n".join("".join(row) for row in out)

# ------------- Agents (optional AI) -------------
class Agent:
    def choose_dir(self, grid, pac_tile):
        return (0, 0)

def _legal_dirs(grid, r, c):
    rows, cols = len(grid), len(grid[0])
    dirs = []
    for dx, dy in [(0,-1),(1,0),(0,1),(-1,0)]:
        nr, nc = r + dy, c + dx
        if 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] != WALL:
            dirs.append((dx, dy))
    return dirs

class RandomAgent(Agent):
    def choose_dir(self, grid, pac_tile):
        r, c = pac_tile
        dirs = _legal_dirs(grid, r, c)
        return random.choice(dirs) if dirs else (0,0)

class GreedyPelletAgent(Agent):
    """BFS to the nearest pellet; returns the FIRST step."""
    def __init__(self, avoid_ghosts=False, ghosts_ref=None):
        self.avoid_ghosts = avoid_ghosts
        self.ghosts_ref = ghosts_ref

    def choose_dir(self, grid, pac_tile):
        r0, c0 = pac_tile

        blocked = set()
        if self.avoid_ghosts and self.ghosts_ref:
            for g in self.ghosts_ref:
                gr, gc = int(g.y // TILE_SIZE), int(g.x // TILE_SIZE)
                blocked.add((gr, gc))

        rows, cols = len(grid), len(grid[0])
        q = deque([(r0, c0)])
        prev = { (r0, c0): None }
        target = None

        while q:
            r, c = q.popleft()
            if grid[r][c] == PELLET and (r, c) != (r0, c0):
                target = (r, c); break
            for dx, dy in [(0,-1),(1,0),(0,1),(-1,0)]:
                nr, nc = r + dy, c + dx
                if 0 <= nr < rows and 0 <= nc < cols and (nr, nc) not in prev:
                    if grid[nr][nc] != WALL and (nr, nc) not in blocked:
                        prev[(nr, nc)] = (r, c)
                        q.append((nr, nc))

        if not target:
            dirs = _legal_dirs(grid, r0, c0)
            return random.choice(dirs) if dirs else (0,0)

        cur = target
        while prev[cur] and prev[cur] != (r0, c0):
            cur = prev[cur]
        r1, c1 = cur
        dx = 0 if c1 == c0 else (1 if c1 > c0 else -1)
        dy = 0 if r1 == r0 else (1 if r1 > r0 else -1)
        return (dx, dy)

# ------------- Entities -------------
class Pacman:
    def __init__(self, r, c):
        self.x = (c + 0.5) * TILE_SIZE
        self.y = (r + 0.5) * TILE_SIZE
        self.dir_x = 0
        self.dir_y = 0
        self.next_dx = 0
        self.next_dy = 0
        self.speed = 2.0  # must divide TILE_SIZE nicely (20 % 2 == 0)
        self.radius = TILE_SIZE // 2 - 2
        self._EPS_CENTER = 0.5
        self.controller = None  # optional AI controller

    def set_controller(self, controller):
        self.controller = controller

    def tile_pos(self):
        return int(self.y // TILE_SIZE), int(self.x // TILE_SIZE)

    def _center_of(self, r, c):
        return (c + 0.5) * TILE_SIZE, (r + 0.5) * TILE_SIZE

    def _is_centered(self):
        cx = (self.x % TILE_SIZE) - TILE_SIZE/2
        cy = (self.y % TILE_SIZE) - TILE_SIZE/2
        return abs(cx) <= self._EPS_CENTER and abs(cy) <= self._EPS_CENTER

    def _snap_to_center(self):
        r, c = self.tile_pos()
        self.x, self.y = self._center_of(r, c)

    def _can_move(self, grid, dx, dy):
        if dx == 0 and dy == 0:
            return True
        rows, cols = len(grid), len(grid[0])
        r, c = self.tile_pos()
        nr, nc = r + dy, c + dx
        return 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] != WALL

    def set_desired_dir(self, dx, dy, grid):
        self.next_dx, self.next_dy = dx, dy
        if self._is_centered() and self._can_move(grid, dx, dy):
            self.dir_x, self.dir_y = dx, dy

    def _advance_towards_next_center(self):
        if self.dir_x == 0 and self.dir_y == 0:
            return
        r, c = self.tile_pos()
        target_r = r + self.dir_y
        target_c = c + self.dir_x
        tx, ty = self._center_of(target_r, target_c)
        vx, vy = tx - self.x, ty - self.y
        step = self.speed

        if vx*vx + vy*vy <= step*step:
            self.x, self.y = tx, ty
            return

        if self.dir_x != 0:
            self.x = min(self.x + step, tx) if self.dir_x > 0 else max(self.x - step, tx)
            self.y = (r + 0.5) * TILE_SIZE
        else:
            self.y = min(self.y + step, ty) if self.dir_y > 0 else max(self.y - step, ty)
            self.x = (c + 0.5) * TILE_SIZE

    def update(self, grid):
        # If we have an AI controller and are centered, let it pick a direction
        if self._is_centered() and self.controller:
            self._snap_to_center()
            dx, dy = self.controller.choose_dir(grid, self.tile_pos())
            self.set_desired_dir(dx, dy, grid)

        if self._is_centered():
            self._snap_to_center()

            if (self.next_dx, self.next_dy) != (self.dir_x, self.dir_y):
                if self._can_move(grid, self.next_dx, self.next_dy):
                    self.dir_x, self.dir_y = self.next_dx, self.next_dy

            if not self._can_move(grid, self.dir_x, self.dir_y):
                self.dir_x, self.dir_y = 0, 0

        self._advance_towards_next_center()

        # Eat pellets only when centered on the tile
        if self._is_centered():
            self._snap_to_center()
            r, c = self.tile_pos()
            if grid[r][c] == PELLET:
                grid[r][c] = EMPTY

    def draw(self, screen):
        angle = 0
        if   self.dir_x < 0: angle = 180
        elif self.dir_x > 0: angle = 0
        elif self.dir_y < 0: angle = 90
        elif self.dir_y > 0: angle = 270
        mouth = 40
        sa = (angle + mouth) % 360
        ea = (angle - mouth) % 360
        pygame.draw.circle(screen, PACMAN_C, (int(self.x), int(self.y)), self.radius)
        v = pygame.math.Vector2
        pts = [
            (self.x, self.y),
            (self.x + self.radius * v(1,0).rotate(-sa).x,
             self.y + self.radius * v(1,0).rotate(-sa).y),
            (self.x + self.radius * v(1,0).rotate(-ea).x,
             self.y + self.radius * v(1,0).rotate(-ea).y),
        ]
        pygame.draw.polygon(screen, BLACK, pts)

class Ghost:
    def __init__(self, r, c, sprite):
        self.x = (c + 0.5) * TILE_SIZE
        self.y = (r + 0.5) * TILE_SIZE
        self.dir_x = 0
        self.dir_y = 0
        self.speed = 2.0
        self.sprite = sprite
        self._EPS_CENTER = 0.5

    def _is_centered(self):
        cx = (self.x % TILE_SIZE) - TILE_SIZE/2
        cy = (self.y % TILE_SIZE) - TILE_SIZE/2
        return abs(cx) <= self._EPS_CENTER and abs(cy) <= self._EPS_CENTER

    def _snap_to_center(self):
        r, c = int(self.y // TILE_SIZE), int(self.x // TILE_SIZE)
        self.x = (c + 0.5) * TILE_SIZE
        self.y = (r + 0.5) * TILE_SIZE

    def _can_move(self, grid, dx, dy):
        rows, cols = len(grid), len(grid[0])
        r, c = int(self.y // TILE_SIZE), int(self.x // TILE_SIZE)
        nr, nc = r + dy, c + dx
        return 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] != WALL

    def _choose_dir(self, grid, pac_r, pac_c):
        col = int(self.x // TILE_SIZE)
        row = int(self.y // TILE_SIZE)
        best_d, best = 1e9, (0,0)
        for dx, dy in [(0,-1),(1,0),(0,1),(-1,0)]:
            nr, nc = row + dy, col + dx
            if self._can_move(grid, dx, dy):
                d = abs(nc - pac_c) + abs(nr - pac_r)
                if d < best_d:
                    best_d, best = d, (dx, dy)
        self.dir_x, self.dir_y = best

    def _advance_towards_next_center(self):
        if self.dir_x == 0 and self.dir_y == 0:
            return
        r, c = int(self.y // TILE_SIZE), int(self.x // TILE_SIZE)
        tx = (c + 0.5 + self.dir_x) * TILE_SIZE
        ty = (r + 0.5 + self.dir_y) * TILE_SIZE
        step = self.speed
        if self.dir_x != 0:
            self.x = min(self.x + step, tx) if self.dir_x > 0 else max(self.x - step, tx)
            self.y = (r + 0.5) * TILE_SIZE
        else:
            self.y = min(self.y + step, ty) if self.dir_y > 0 else max(self.y - step, ty)
            self.x = (c + 0.5) * TILE_SIZE

    def update(self, grid, pac_r, pac_c):
        if self._is_centered():
            self._snap_to_center()
            self._choose_dir(grid, pac_r, pac_c)
            if not self._can_move(grid, self.dir_x, self.dir_y):
                self.dir_x, self.dir_y = 0, 0
        self._advance_towards_next_center()

    def draw(self, screen):
        if self.sprite and hasattr(self.sprite, "get_rect"):
            rect = self.sprite.get_rect(center=(int(self.x), int(self.y)))
            screen.blit(self.sprite, rect)
        else:
            pygame.draw.circle(screen, (255,0,0), (int(self.x), int(self.y)), TILE_SIZE//2 - 2)

# ------------- Game Setup -------------
def make_game_from_text(txt):
    grid, pac_spawn, ghost_spawns = load_map_from_text(txt)
    rows, cols = len(grid), len(grid[0])
    width, height = cols*TILE_SIZE, rows*TILE_SIZE

    pac = Pacman(*pac_spawn)
    ghost_sprite = None
    if os.path.exists(GHOST_SPRITE):
        try:
            ghost_sprite = pygame.image.load(GHOST_SPRITE).convert_alpha()
            ghost_sprite = pygame.transform.scale(ghost_sprite, (TILE_SIZE, TILE_SIZE))
        except Exception:
            ghost_sprite = None
    ghosts = [Ghost(r, c, ghost_sprite) for (r, c) in ghost_spawns]
    return grid, pac, ghosts, pac_spawn, ghost_spawns, rows, cols, width, height

def count_pellets(g):
    return sum(cell == PELLET for row in g for cell in row)

# ------------- Main -------------
def main():
    pygame.init()
    grid, pac, ghosts, pac_spawn, ghost_spawns, ROWS, COLS, WIDTH, HEIGHT = make_game_from_text(MAP_TEXT)
    screen = pygame.display.set_mode((WIDTH, HEIGHT))
    pygame.display.set_caption("Pac-Man (inline ASCII map)")
    clock = pygame.time.Clock()
    font = pygame.font.SysFont(None, 18)

    editing = False
    paint_tile = WALL
    pellets_left = count_pellets(grid)

    # --- Optional AI ---
    ai_enabled = True
    agent = GreedyPelletAgent(avoid_ghosts=True, ghosts_ref=ghosts)
    pac.set_controller(agent if ai_enabled else None)

    def reset_entities():
        nonlocal pac, ghosts, pellets_left
        spr = ghosts[0].sprite if ghosts else None
        pac = Pacman(*pac_spawn)
        pac.set_controller(agent if ai_enabled else None)
        ghosts[:] = [Ghost(r, c, spr) for (r, c) in ghost_spawns]
        pellets_left = count_pellets(grid)

    running = True
    while running:
        dt = clock.tick(FPS)
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

            elif event.type == pygame.KEYDOWN:
                # Toggle AI
                if event.key == pygame.K_a:
                    ai_enabled = not ai_enabled
                    pac.set_controller(agent if ai_enabled else None)

                # Player input only when AI is OFF
                if not ai_enabled:
                    if event.key == pygame.K_UP:    pac.set_desired_dir(0, -1, grid)
                    if event.key == pygame.K_DOWN:  pac.set_desired_dir(0,  1, grid)
                    if event.key == pygame.K_LEFT:  pac.set_desired_dir(-1, 0, grid)
                    if event.key == pygame.K_RIGHT: pac.set_desired_dir(1,  0, grid)

                # Editor
                if event.key == pygame.K_e: editing = not editing
                if event.key == pygame.K_1: paint_tile = WALL
                if event.key == pygame.K_2: paint_tile = PELLET
                if event.key == pygame.K_3: paint_tile = EMPTY
                if event.key == pygame.K_p: # set Pac spawn at mouse tile
                    mx, my = pygame.mouse.get_pos()
                    c, r = mx//TILE_SIZE, my//TILE_SIZE
                    pac_spawn = (r, c); reset_entities()
                if event.key == pygame.K_g: # add a Ghost spawn
                    mx, my = pygame.mouse.get_pos()
                    c, r = mx//TILE_SIZE, my//TILE_SIZE
                    ghost_spawns.append((r, c)); reset_entities()
                if event.key == pygame.K_r: reset_entities()
                if event.key == pygame.K_s:
                    print("\n----- COPY BELOW INTO MAP_TEXT -----")
                    print(stringify_map(grid, pac_spawn, ghost_spawns))
                    print("-----  COPY END  -----\n")

            elif event.type == pygame.MOUSEBUTTONDOWN and editing:
                mx, my = event.pos
                c, r = mx//TILE_SIZE, my//TILE_SIZE
                if 0 <= r < ROWS and 0 <= c < COLS:
                    grid[r][c] = paint_tile
                    pellets_left = count_pellets(grid)

        if not editing:
            pac.update(grid)
            pr, pc = pac.tile_pos()
            pellets_left = count_pellets(grid)
            for g in ghosts:
                g.update(grid, pr, pc)
                gr, gc = int(g.y//TILE_SIZE), int(g.x//TILE_SIZE)
                if (gr, gc) == (pr, pc):
                    reset_entities()
                    break

        # Win: all pellets eaten -> refill pellets and continue
        if pellets_left == 0 and not editing:
            for r in range(ROWS):
                for c in range(COLS):
                    if grid[r][c] == EMPTY:
                        grid[r][c] = PELLET
            pellets_left = count_pellets(grid)
            reset_entities()

        # Draw
        screen.fill(BLACK)
        for r in range(ROWS):
            for c in range(COLS):
                cell = grid[r][c]
                if cell == WALL:
                    pygame.draw.rect(screen, WALL_C, (c*TILE_SIZE, r*TILE_SIZE, TILE_SIZE, TILE_SIZE))
                elif cell == PELLET:
                    pygame.draw.circle(screen, PELLET_C,
                        (c*TILE_SIZE + TILE_SIZE//2, r*TILE_SIZE + TILE_SIZE//2), 3)

        if editing:
            for r in range(ROWS):
                pygame.draw.line(screen, GRID_C, (0, r*TILE_SIZE), (WIDTH, r*TILE_SIZE), 1)
            for c in range(COLS):
                pygame.draw.line(screen, GRID_C, (c*TILE_SIZE, 0), (c*TILE_SIZE, HEIGHT), 1)
            mx, my = pygame.mouse.get_pos()
            hc, hr = mx//TILE_SIZE, my//TILE_SIZE
            pygame.draw.rect(screen, EDIT_HI, (hc*TILE_SIZE, hr*TILE_SIZE, TILE_SIZE, TILE_SIZE), 1)

        pac.draw(screen)
        for g in ghosts:
            g.draw(screen)

        # HUD
        hud_lines = [
            f"Pellets: {pellets_left}",
            "[A] AI: " + ("ON" if ai_enabled else "OFF") + "   [E] Edit   [R] Reset   [S] Print Map",
            "[Arrows] Move (AI OFF)   [1] Wall  [2] Pellet  [3] Empty   [P] Set Pac  [G] Add Ghost",
        ]
        y = 4
        for line in hud_lines:
            screen.blit(font.render(line, True, (200,200,200)), (6, y))
            y += 18

        pygame.display.flip()

    pygame.quit(); sys.exit()

if __name__ == "__main__":
    main()
```

```python
# pacman_inline_map.py
import pygame, sys, os
from collections import deque

# =========================
# Inline ASCII Map (edit me)
# =========================
# MAP_TEXT = """
# ############################
# ##########
# #.####.#####.##.#####.####.#
# #P#  #.#   #.##.#   #.#  #.#
# #.####.#####.##.#####.####.#
# ####..............#
# #.####.##.########.##.####.#
# #G   #.##....##....##.#   G#
# #### #.##### ## #####.# ####
#      #.##### ## #####.#     
# #### #.##          ##.# ####
# #    .## ###GGGG### ##.    #
# #### .## #        # ## .####
# #.....   #        #   .....#
# #### .## #        # ## .####
# #    .## ######## ## .    #
# #### #.##### ## #####.# ####
#      #.##### ## #####.#     
# #### #.##          ##.# ####
# #G   #.##....##....##.#   G#
# #.####.##.########.##.####.#
# #..........................#
# #.####.#####.##.#####.####.#
# #.#  #.#   #.##.#   #.#  #.#
# #.####.#####.##.#####.####.#
# #............##............#
# ############################
# """.strip("\n")

MAP_TEXT = """
####################
#........P.........#
#....###.....###...#
#..................#
#..................#
#....###.....###...#
#..................#
#..................#
#....###.....###...#
#....G.........G...#
#..................#
#..................#
####################
""".strip("\n")

# ------------- Config -------------
TILE_SIZE = 20
FPS = 60
GHOST_SPRITE = 'ghost.png'

# Tile codes
EMPTY, WALL, PELLET = 0, 1, 2

# Colours
BLACK   = (0, 0, 0)
WALL_C  = (0, 0, 255)
PELLET_C= (255, 255, 0)
PACMAN_C= (255, 255, 0)
GRID_C  = (40, 40, 40)
EDIT_HI = (60, 60, 60)

CHAR2TILE = {'#': WALL, '.': PELLET, ' ': EMPTY}
TILE2CHAR = {v:k for k,v in CHAR2TILE.items()}

# ------------- Map Parse / Print -------------
def load_map_from_text(txt):
    lines = [line.rstrip('\n') for line in txt.splitlines()]
    rows = len(lines)
    cols = max(len(r) for r in lines)
    lines = [r.ljust(cols, ' ') for r in lines]

    grid = [[EMPTY]*cols for _ in range(rows)]
    pac_spawn = None
    ghost_spawns = []

    for r, row in enumerate(lines):
        for c, ch in enumerate(row):
            if ch in CHAR2TILE:
                grid[r][c] = CHAR2TILE[ch]
            elif ch == 'P':
                pac_spawn = (r, c)
            elif ch == 'G':
                ghost_spawns.append((r, c))
            else:
                grid[r][c] = EMPTY

    # If no pellets were in the text, auto-fill pellets on empties
    if not any(PELLET in row for row in grid):
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == EMPTY:
                    grid[r][c] = PELLET

    if pac_spawn is None:
        pac_spawn = (1, 1)
    if not ghost_spawns:
        ghost_spawns = [(rows-2, cols-2)]

    return grid, pac_spawn, ghost_spawns

def stringify_map(grid, pac_spawn, ghost_spawns):
    rows, cols = len(grid), len(grid[0])
    out = [[' ']*cols for _ in range(rows)]
    for r in range(rows):
        for c in range(cols):
            out[r][c] = TILE2CHAR.get(grid[r][c], ' ')
    pr, pc = pac_spawn
    out[pr][pc] = 'P'
    for gr, gc in ghost_spawns:
        out[gr][gc] = 'G'
    return "\n".join("".join(row) for row in out)

# ------------- Entities -------------
class Pacman:
    def __init__(self, r, c):
        self.x = (c + 0.5) * TILE_SIZE
        self.y = (r + 0.5) * TILE_SIZE
        self.dir_x = 0
        self.dir_y = 0
        self.speed = 2.0
        self.radius = TILE_SIZE//2 - 2

    def tile_pos(self):
        return int(self.y // TILE_SIZE), int(self.x // TILE_SIZE)

    def center_offset(self):
        cx = (self.x % TILE_SIZE) - TILE_SIZE/2
        cy = (self.y % TILE_SIZE) - TILE_SIZE/2
        return cx, cy

    def update(self, grid):
        rows, cols = len(grid), len(grid[0])
        nx = self.x + self.dir_x*self.speed
        ny = self.y + self.dir_y*self.speed
        col = int(max(0, min(cols-1, nx//TILE_SIZE)))
        row = int(max(0, min(rows-1, ny//TILE_SIZE)))
        if grid[row][col] != WALL:
            self.x, self.y = nx, ny

        row, col = self.tile_pos()
        if 0 <= row < rows and 0 <= col < cols and grid[row][col] == PELLET:
            cx, cy = self.center_offset()
            if abs(cx) < 6 and abs(cy) < 6:
                grid[row][col] = EMPTY

    def draw(self, screen):
        angle = 0
        if   self.dir_x < 0: angle = 180
        elif self.dir_x > 0: angle = 0
        elif self.dir_y < 0: angle = 90
        elif self.dir_y > 0: angle = 270
        mouth = 40
        sa = (angle + mouth) % 360
        ea = (angle - mouth) % 360
        pygame.draw.circle(screen, PACMAN_C, (int(self.x), int(self.y)), self.radius)
        v = pygame.math.Vector2
        pts = [
            (self.x, self.y),
            (self.x + self.radius * v(1,0).rotate(-sa).x,
             self.y + self.radius * v(1,0).rotate(-sa).y),
            (self.x + self.radius * v(1,0).rotate(-ea).x,
             self.y + self.radius * v(1,0).rotate(-ea).y),
        ]
        # pygame.draw.polygon(screen, BLACK, pts) # Cuts a triangle into the pacman mouth

class Ghost:
    def __init__(self, r, c, sprite):
        self.x = (c + 0.5) * TILE_SIZE
        self.y = (r + 0.5) * TILE_SIZE
        self.dir_x = 0
        self.dir_y = 0
        self.speed = 2.0
        self.sprite = sprite

    def _choose_dir(self, grid, pac_r, pac_c):
        rows, cols = len(grid), len(grid[0])
        col = int(self.x // TILE_SIZE)
        row = int(self.y // TILE_SIZE)
        best_d, best = 1e9, (0,0)
        for dx, dy in [(0,-1),(1,0),(0,1),(-1,0)]:
            nr, nc = row + dy, col + dx
            if 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] != WALL:
                d = abs(nc - pac_c) + abs(nr - pac_r)
                if d < best_d:
                    best_d, best = d, (dx, dy)
        self.dir_x, self.dir_y = best

    def update(self, grid, pac_r, pac_c):
        if abs((self.x % TILE_SIZE) - TILE_SIZE/2) < 5 and abs((self.y % TILE_SIZE) - TILE_SIZE/2) < 5:
            self._choose_dir(grid, pac_r, pac_c)
        rows, cols = len(grid), len(grid[0])
        nx = self.x + self.dir_x*self.speed
        ny = self.y + self.dir_y*self.speed
        col = int(max(0, min(cols-1, nx//TILE_SIZE)))
        row = int(max(0, min(rows-1, ny//TILE_SIZE)))
        if grid[row][col] != WALL:
            self.x, self.y = nx, ny
        else:
            self.dir_x = self.dir_y = 0

    def draw(self, screen):
        if self.sprite and hasattr(self.sprite, "get_rect"):
            rect = self.sprite.get_rect(center=(int(self.x), int(self.y)))
            screen.blit(self.sprite, rect)
        else:
            pygame.draw.circle(screen, (255,0,0), (int(self.x), int(self.y)), TILE_SIZE//2 - 2)

# ------------- Game Setup -------------
def make_game_from_text(txt):
    grid, pac_spawn, ghost_spawns = load_map_from_text(txt)
    rows, cols = len(grid), len(grid[0])
    width, height = cols*TILE_SIZE, rows*TILE_SIZE

    pac = Pacman(*pac_spawn)
    ghost_sprite = None
    if os.path.exists(GHOST_SPRITE):
        ghost_sprite = pygame.image.load(GHOST_SPRITE).convert_alpha()
        ghost_sprite = pygame.transform.scale(ghost_sprite, (TILE_SIZE, TILE_SIZE))
    ghosts = [Ghost(r, c, ghost_sprite) for (r, c) in ghost_spawns]
    return grid, pac, ghosts, pac_spawn, ghost_spawns, rows, cols, width, height

def count_pellets(g):
    return sum(cell == PELLET for row in g for cell in row)

# ------------- Main -------------
def main():
    pygame.init() 
    grid, pac, ghosts, pac_spawn, ghost_spawns, ROWS, COLS, WIDTH, HEIGHT = make_game_from_text(MAP_TEXT)
    screen = pygame.display.set_mode((WIDTH, HEIGHT))
    pygame.display.set_caption("Pac-Man (inline ASCII map)")
    clock = pygame.time.Clock()
    font = pygame.font.SysFont(None, 18)

    editing = False
    paint_tile = WALL
    pellets_left = count_pellets(grid)

    def reset_entities():
        nonlocal pac, ghosts, pellets_left
        # keep map as-is
        spr = ghosts[0].sprite if ghosts else None
        pac = Pacman(*pac_spawn)
        ghosts[:] = [Ghost(r, c, spr) for (r, c) in ghost_spawns]
        pellets_left = count_pellets(grid)

    reset_entities()

    running = True
    while running:
        dt = clock.tick(FPS)
        for event in pygame.event.get():
            if event.type == pygame.QUIT: running = False
            elif event.type == pygame.KEYDOWN:
                if event.key == pygame.K_UP:    pac.dir_x, pac.dir_y = 0, -1
                if event.key == pygame.K_DOWN:  pac.dir_x, pac.dir_y = 0,  1
                if event.key == pygame.K_LEFT:  pac.dir_x, pac.dir_y = -1, 0
                if event.key == pygame.K_RIGHT: pac.dir_x, pac.dir_y = 1,  0

                # Editor
                if event.key == pygame.K_e: editing = not editing
                if event.key == pygame.K_1: paint_tile = WALL
                if event.key == pygame.K_2: paint_tile = PELLET
                if event.key == pygame.K_3: paint_tile = EMPTY
                if event.key == pygame.K_p: # set Pac spawn at mouse tile
                    mx, my = pygame.mouse.get_pos()
                    c, r = mx//TILE_SIZE, my//TILE_SIZE
                    pac_spawn = (r, c); reset_entities()
                if event.key == pygame.K_g: # add a Ghost spawn
                    mx, my = pygame.mouse.get_pos()
                    c, r = mx//TILE_SIZE, my//TILE_SIZE
                    ghost_spawns.append((r, c)); reset_entities()
                if event.key == pygame.K_r: reset_entities()
                if event.key == pygame.K_s:
                    # Print the current map to console so you can paste back into MAP_TEXT
                    print("\n----- COPY BELOW INTO MAP_TEXT -----")
                    print(stringify_map(grid, pac_spawn, ghost_spawns))
                    print("-----  COPY END  -----\n")

            elif event.type == pygame.MOUSEBUTTONDOWN and editing:
                mx, my = event.pos
                c, r = mx//TILE_SIZE, my//TILE_SIZE
                if 0 <= r < ROWS and 0 <= c < COLS:
                    grid[r][c] = paint_tile
                    pellets_left = count_pellets(grid)

        if not editing:
            pac.update(grid)
            pr, pc = pac.tile_pos()
            pellets_left = count_pellets(grid)
            for g in ghosts:
                g.update(grid, pr, pc)
                gr, gc = int(g.y//TILE_SIZE), int(g.x//TILE_SIZE)
                if (gr, gc) == (pr, pc):
                    reset_entities()
                    break

        # Win: all pellets eaten -> refill pellets and continue
        if pellets_left == 0 and not editing:
            for r in range(ROWS):
                for c in range(COLS):
                    if grid[r][c] == EMPTY:
                        grid[r][c] = PELLET
            pellets_left = count_pellets(grid)
            reset_entities()

        # Draw
        screen.fill(BLACK)
        for r in range(ROWS):
            for c in range(COLS):
                cell = grid[r][c]
                if cell == WALL:
                    pygame.draw.rect(screen, WALL_C, (c*TILE_SIZE, r*TILE_SIZE, TILE_SIZE, TILE_SIZE))
                elif cell == PELLET:
                    pygame.draw.circle(screen, PELLET_C,
                        (c*TILE_SIZE + TILE_SIZE//2, r*TILE_SIZE + TILE_SIZE//2), 3)

        if editing:
            for r in range(ROWS):
                pygame.draw.line(screen, GRID_C, (0, r*TILE_SIZE), (WIDTH, r*TILE_SIZE), 1)
            for c in range(COLS):
                pygame.draw.line(screen, GRID_C, (c*TILE_SIZE, 0), (c*TILE_SIZE, HEIGHT), 1)
            mx, my = pygame.mouse.get_pos()
            hc, hr = mx//TILE_SIZE, my//TILE_SIZE
            pygame.draw.rect(screen, EDIT_HI, (hc*TILE_SIZE, hr*TILE_SIZE, TILE_SIZE, TILE_SIZE), 1)

        pac.draw(screen)
        for g in ghosts:
            g.draw(screen)

        # HUD
        hud_lines = [
            f"Pellets: {pellets_left}",
            "[Arrows] Move   [R] Reset   [E] Edit/Play   [S] Print Map",
            "[1] Wall  [2] Pellet  [3] Empty   [P] Set Pac  [G] Add Ghost",
        ]
        font = pygame.font.SysFont(None, 18)
        y = 4
        for line in hud_lines:
            screen.blit(font.render(line, True, (200,200,200)), (6, y))
            y += 18

        pygame.display.flip()

    pygame.quit(); sys.exit()

if __name__ == "__main__":
    main()
```
